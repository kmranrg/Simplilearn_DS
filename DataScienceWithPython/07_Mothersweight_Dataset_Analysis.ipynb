```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Mothersweight.xlsx")
df
```

```python
x = df['Mothers_weight']
y = df['Infants_birthweight']
```

```python
plt.scatter(x,y)
plt.xlabel('Mothers_weight')
plt.ylabel('Infants_birthweight')
plt.show()
```

```python
'''
Analysis

1) We can see from above graph that there is no correlation between the Mothers_weight and Infants_birthweight
'''
```

```python
# let's calculate the r value

from scipy.stats import pearsonr
corr, _ = pearsonr(x,y)
print('Pearson correlation',corr)
```

```python
df.corr()
```

```python
'''
from above table also we can see that correlation between mothers weight and infants birthweight is 0.521
'''
```

```python
'''
Analysis:

1) Based on r value - correlation is fair and moderate
2) It means mother's weight is not the only parameter which depends on infant's weight (only 52% depends)
3) There are some other parameters as well on which the infant's weight depends like mother's food habit, exercise, etc.
'''
```
