## Data Preprocessing

```python
df_required = df.drop('Car_Name', axis=1) # we are removing the Car_Name feature, as it's not required in predicting the Selling_Price (told by Domain Expert)
df_required
```

```python
# adding the present year feature in the dataset
df_required['Present_Year'] = 2023
```

```python
df_required
```

```python
df_required['Age_Year'] = df_required['Present_Year'] - df_required['Year']
df_required
```

```python
# as now we have the age of car, so now 'Year' & 'Present_Year' features are not required
df_required.drop(['Year','Present_Year'], axis=1, inplace=True)
df_required
```

```python
df_required.dtypes
```

```python
# conversion of object data types to integers
df_required = pd.get_dummies(df_required)
df_required
```
