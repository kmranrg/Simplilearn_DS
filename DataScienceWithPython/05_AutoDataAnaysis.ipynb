```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Auto.csv')
df
```

```python
'''
mpg - mileage
weight - weight of the vehicle
'''
```

```python
x = df['weight']
y = df['mpg']
```

```python
plt.scatter(x,y)
plt.title('Mileage vs Weight of Vehicle')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
df.corr()
```

```python
'''
Analysis

1) r value = -0.87
2) mpg and weight are inversely proportional
3) Strong negative correlation
'''
```

### Manual calculation of correlation coefficient (r)

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(x,y)
print('Pearson Correlation:',corr)
```
