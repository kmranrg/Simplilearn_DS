```python
df = df.drop(['id','diagnosis','Unnamed: 32'], axis=1)
```

```python
df.describe()
```

## Standardizing the Features Data

```python
from sklearn.preprocessing import StandardScaler
```

<img src='25.png'>

```python
standardized = StandardScaler()
```

```python
standardized.fit(df)
```

```python
scaled_data = pd.DataFrame(standardized.transform(df))
```

```python
scaled_data
```

```python
scaled_data.describe()
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit(scaled_data) # now using fit, we are checking how many PCAs are needed for our 30 featured dataset
```

```python
vr = pca.explained_variance_ratio_ # variance
```

```python
vr.shape
```

```python
import numpy as np

# cumulative variance
cum_vr = np.cumsum(vr)
```

```python
plt.plot(cum_vr)
```

<img src='26.png'>

```python
pca = PCA(n_components=15)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
x_pca
```
