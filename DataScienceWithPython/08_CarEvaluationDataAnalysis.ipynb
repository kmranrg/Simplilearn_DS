```python
import pandas as pd
```

```python
data = pd.read_csv('car_evaluation.csv')
data
```

```python
'''
In the above dataset, headers are missing, that means column names are not present,
so let's sort this problem, let's add the header.

Note: By default, pandas will take the first row as header
'''
```

```python
columns = ['buying','maintenance','doors','person_capacity','lug_boot','safety','cars']
```

```python
data.columns = columns
```

```python
data
```

```python
'''
We see there are some columns with string values.
Let's convert them to integers withour using get_dummies()
'''
```

```python
## Converting the data type string to integer

buying = {'vhigh':10,'high':8,'med':5,'low':3}
data.buying = [buying[i] for i in data.buying ]

maintenance = {'vhigh':200,'high':150,'med':100,'low':50}
data.maintenance = [maintenance[j] for j in data.maintenance ]

doors = {'2':2,'3':3,'4':4,'5more':6}
data.doors = [doors[j] for j in data.doors ]

persons = {'2':2,'4':4,'more':5}
data.person_capacity = [persons[j] for j in data.person_capacity ]

lug_boot = {'small':250,'med':500,'big':800}
data.lug_boot = [lug_boot[j] for j in data.lug_boot ]

safety = {'high':5,'med':4,'low':3}
data.safety = [safety[j] for j in data.safety ]

cars = {'vgood':1,'good':2,'acc':3,'unacc':4}
data.cars = [cars[j] for j in data.cars ]

data
```
