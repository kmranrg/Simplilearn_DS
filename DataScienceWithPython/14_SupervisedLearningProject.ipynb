# <center>Salary DataSet</center>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Salary_Data.csv')
data
```

### 1. Determining Features(x) and label(y)

```python
x = pd.DataFrame(data['YearsExperience'])
y = pd.DataFrame(data['Salary'])
```

```python
x
```

```python
y
```

### 2. Splitting the dataset into Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

<img src="8.png">

### 3. Training the Algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

### 4. Predicting the test data (x_test)

```python
y_pred = regressor.predict(x_test) # testing my model
y_pred # predicted output
```

```python
# actual output
y_test
```

### 5. Evaluating the model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# r-sqaured value
r2 = r2_score(y_test, y_pred)
print(r2*100)
```

<img src="10.png">

```python
print('Mean absolute error', mean_absolute_error(y_test, y_pred))
```

```python
print('Mean squared error', mean_squared_error(y_test, y_pred))
```

```python
from math import sqrt
print('RMSE',sqrt(mean_absolute_error(y_test, y_pred)))
```
