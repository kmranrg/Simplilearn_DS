```python
'''
1) Positive Correlation - when x increases, y also increases
2) Negative Correlation - when x increases, y decreases
3) Thickness also matters in correlation graph - less thickness --> strong correlation & more thickness - weak correlation
'''
```

```python
import pandas as pd # to handle the dataset
import matplotlib.pyplot as plt # for visualization
```

```python
df = pd.read_csv('Woodenboard.csv')
df
```

```python
'''
x -> density
y -> stiffness
'''
```

```python
x = df['Density']
y = df['Stiffness']
```

```python
plt.scatter(x,y)
plt.title('Wooden Board')
plt.xlabel('density')
plt.ylabel('stiffness')
plt.show()
```

```python
df['test'] = df['Density'] + 100
df
```

```python
df = df.drop('test',axis=1)
df
```

```python
df.corr()
```

```python
'''
Analysis from above cell:

As the r-value is 0.91 - that means strong correlation
'''
```
