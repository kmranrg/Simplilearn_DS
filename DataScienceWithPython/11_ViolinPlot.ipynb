```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('gapminder_full.csv')
data
```

```python
country = data.country
year = data.year
population = data.population
continent = data.continent
life_exp = data.life_exp
gdp_cap = data.gdp_cap
```

```python
sns.violinplot(x=life_exp)
plt.show()
```

```python
sns.violinplot(x=continent, y=life_exp)
plt.show()
```

```python
'''
Violin plot adds density to the box plot
'''
```

```python
'''
Limitation of Violin Plot

1) It does not tell the frequency of the dataset
'''
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x=continent, y=life_exp)
plt.show()
```

```python
'''
Swarm plot shows the frequency of dataset
'''
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x=continent, y=life_exp)
sns.swarmplot(x=continent, y=life_exp)
plt.show()
```
